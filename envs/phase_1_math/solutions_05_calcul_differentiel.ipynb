## ğŸ“ˆ Section 2 : DÃ©rivÃ©es Basiques - Solutions

```python
print('Solution 2.1: DÃ©rivÃ©e avec DÃ©finition')
print('='*60)
h = sp.Symbol('h')
f = x**2
diff_q = (f.subs(x, x+h) - f) / h
derivee = sp.limit(diff_q, h, 0)
print(f'f(x) = xÂ²')
print(f"f\'(x) = {derivee}")
print(f'DÃ©rivÃ©e par dÃ©finition confirmÃ©e!')

print('\nSolution 2.2: DÃ©rivÃ©es de PolynÃ´mes')
print('='*60)
polys = [3*x**4 - 2*x**3 + 5*x - 7, x**5 + 6*x**2 - 9, (2*x+1)**3, x**3/3 - x**2/2 + x]
for i, p in enumerate(polys, 1):
    print(f'{i}. f(x)={sp.simplify(p)} â†’ f\'(x)={sp.diff(p,x)}')

print('\nSolution 2.3: Pente de la Tangente')
print('='*60)
f = x**3 - 3*x
f_p = sp.diff(f, x)
print(f"f\'(1) = {f_p.subs(x, 1)} (pente en x=1)")
print(f"f\'(-2) = {f_p.subs(x, -2)} (pente en x=-2)")
y_1 = f.subs(x, 1)
m_1 = f_p.subs(x, 1)
print(f'Ã‰quation tangente en x=1: y = {m_1}(x-1) + {y_1} = {m_1}x + {y_1-m_1}')
```
