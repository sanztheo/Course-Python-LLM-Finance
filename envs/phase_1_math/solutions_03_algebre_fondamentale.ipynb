### Solution 2.2 : Angle entre Vecteurs

```python
# Solution 2.2

def angle_entre_vecteurs(u, v):
    """Calcule l'angle en degr√©s entre deux vecteurs."""
    cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
    cos_theta = np.clip(cos_theta, -1, 1)  # √âviter erreurs d'arrondi
    theta_rad = np.arccos(cos_theta)
    return np.degrees(theta_rad)

# Angle 1
u1 = np.array([1, 0])
v1 = np.array([1, 1])
angle1 = angle_entre_vecteurs(u1, v1)
print(f"1) Angle entre {u1} et {v1} = {angle1:.2f}¬∞")
print("   (45¬∞, comme attendu pour un vecteur diagonal)\n")

# Angle 2
u2 = np.array([3, 4])
v2 = np.array([4, 3])
angle2 = angle_entre_vecteurs(u2, v2)
print(f"2) Angle entre {u2} et {v2} = {angle2:.2f}¬∞")
print("   (angle aigu, vecteurs dans des directions similaires)\n")

# Angle 3
u3 = np.array([1, 2])
v3 = np.array([-2, 1])
angle3 = angle_entre_vecteurs(u3, v3)
print(f"3) Angle entre {u3} et {v3} = {angle3:.2f}¬∞")
print("   (90¬∞, ces vecteurs sont orthogonaux !)")
```

### Solution 2.3 : Vecteurs Orthogonaux

```python
# Solution 2.3

def sont_orthogonaux(u, v):
    """V√©rifie si deux vecteurs sont orthogonaux."""
    produit = np.dot(u, v)
    return np.isclose(produit, 0)

# Paire 1
u1 = np.array([1, 0])
v1 = np.array([0, 1])
print(f"1) {u1} et {v1}")
print(f"   Produit scalaire: {np.dot(u1, v1)}")
print(f"   Orthogonaux ? {sont_orthogonaux(u1, v1)}\n")

# Paire 2
u2 = np.array([3, 4])
v2 = np.array([-4, 3])
print(f"2) {u2} et {v2}")
print(f"   Produit scalaire: {np.dot(u2, v2)}")
print(f"   Orthogonaux ? {sont_orthogonaux(u2, v2)}")
print(f"   Calcul: 3√ó(-4) + 4√ó3 = -12 + 12 = 0 ‚úì\n")

# Paire 3
u3 = np.array([1, 1])
v3 = np.array([1, -1])
print(f"3) {u3} et {v3}")
print(f"   Produit scalaire: {np.dot(u3, v3)}")
print(f"   Orthogonaux ? {sont_orthogonaux(u3, v3)}")
print(f"   Calcul: 1√ó1 + 1√ó(-1) = 1 - 1 = 0 ‚úì")
```

### Solution 2.4 : Projection d'un Vecteur

```python
# Solution 2.4

def projection(u, v):
    """Calcule la projection de u sur v."""
    return (np.dot(u, v) / np.dot(v, v)) * v

# Projection 1
u1 = np.array([3, 4])
v1 = np.array([1, 0])
proj1 = projection(u1, v1)
print("1) Projection de [3, 4] sur [1, 0]:")
print(f"   R√©sultat: {proj1}")
print("   Interpr√©tation: Projection sur l'axe x = composante x seulement\n")

# Projection 2
u2 = np.array([2, 3])
v2 = np.array([1, 1])
proj2 = projection(u2, v2)
print("2) Projection de [2, 3] sur [1, 1]:")
print(f"   R√©sultat: {proj2}")
print(f"   Calcul: ((2√ó1 + 3√ó1) / (1√ó1 + 1√ó1)) √ó [1, 1]")
print(f"          = (5/2) √ó [1, 1] = [2.5, 2.5] ‚úì")
```

### Solution 2.5 : Similarit√© Cosinus

```python
# Solution 2.5

def similarite_cosinus(u, v):
    """Calcule la similarit√© cosinus entre deux vecteurs."""
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

# Similarit√© 1
u1 = np.array([1, 2, 3])
v1 = np.array([2, 4, 6])
sim1 = similarite_cosinus(u1, v1)
print("1) Similarit√© entre [1, 2, 3] et [2, 4, 6]:")
print(f"   R√©sultat: {sim1:.4f}")
print("   = 1 ‚Üí Vecteurs identiques (m√™me direction) ‚úì\n")

# Similarit√© 2
u2 = np.array([1, 0])
v2 = np.array([0, 1])
sim2 = similarite_cosinus(u2, v2)
print("2) Similarit√© entre [1, 0] et [0, 1]:")
print(f"   R√©sultat: {sim2:.4f}")
print("   = 0 ‚Üí Vecteurs orthogonaux (perpendiculaires) ‚úì\n")

# Similarit√© 3
u3 = np.array([1, 1])
v3 = np.array([-1, -1])
sim3 = similarite_cosinus(u3, v3)
print("3) Similarit√© entre [1, 1] et [-1, -1]:")
print(f"   R√©sultat: {sim3:.4f}")
print("   = -1 ‚Üí Vecteurs oppos√©s (directions contraires) ‚úì")
```

## 5Ô∏è‚É£ Section 5 : Transposition et Inverse - Solutions

### Solution 5.1 : Transposition

### Solution 5.2 : Propri√©t√© $(AB)^T = B^T A^T$

### Solution 6.1 : Calculer le D√©terminant

```python
# Solution 6.1

# Matrice A
A = np.array([[3, 2],
              [1, 4]])
det_A = np.linalg.det(A)
det_A_manuel = A[0,0]*A[1,1] - A[0,1]*A[1,0]
print("Matrice A:")
print(A)
print(f"\nD√©terminant (NumPy): {det_A}")
print(f"D√©terminant (manuel): 3√ó4 - 2√ó1 = 12 - 2 = {det_A_manuel}")
print()

# Matrice B (diagonale)
B = np.array([[5, 0],
              [0, 3]])
det_B = np.linalg.det(B)
print("Matrice B (diagonale):")
print(B)
print(f"\nD√©terminant: {det_B}")
print("Pour une matrice diagonale: det = produit des √©l√©ments diagonaux")
print(f"5 √ó 3 = {5*3} ‚úì")
print()

# Matrice C (triangulaire)
C = np.array([[1, 2, 3],
              [0, 4, 5],
              [0, 0, 6]])
det_C = np.linalg.det(C)
print("Matrice C (triangulaire):")
print(C)
print(f"\nD√©terminant: {det_C}")
print("Pour une matrice triangulaire: det = produit de la diagonale")
print(f"1 √ó 4 √ó 6 = {1*4*6} ‚úì")
```

### Solution 6.2 : D√©terminant et Inversibilit√©

```python
# Solution 6.2

matrices = [
    ("A", np.array([[1, 2], [3, 4]])),
    ("B", np.array([[2, 4], [1, 2]])),
    ("C", np.eye(3))
]

for nom, mat in matrices:
    det = np.linalg.det(mat)
    inversible = not np.isclose(det, 0)
    
    print(f"Matrice {nom}:")
    print(mat)
    print(f"D√©terminant: {det:.6f}")
    print(f"Inversible ? {inversible}")
    
    if inversible:
        print("‚úì det ‚â† 0 ‚Üí matrice inversible")
    else:
        print("‚úó det = 0 ‚Üí matrice NON inversible (singuli√®re)")
    print()
```

### Solution 6.3 : Propri√©t√© $\det(AB) = \det(A)\det(B)$

```python
# Solution 6.3

A = np.array([[2, 1],
              [1, 3]])
B = np.array([[4, 2],
              [1, 5]])

# M√©thode 1: det(AB)
AB = A @ B
det_AB = np.linalg.det(AB)
print("M√©thode 1: det(AB):")
print(f"det(AB) = {det_AB}")
print()

# M√©thode 2: det(A) √ó det(B)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
produit_det = det_A * det_B
print("M√©thode 2: det(A) √ó det(B):")
print(f"det(A) = {det_A}")
print(f"det(B) = {det_B}")
print(f"det(A) √ó det(B) = {produit_det}")
print()

# V√©rification
print("Sont-ils √©gaux (avec tol√©rance) ?")
print(np.isclose(det_AB, produit_det))
print("Oui ! La propri√©t√© det(AB) = det(A)√ódet(B) est v√©rifi√©e. ‚úì")
```

### Solution 6.4 : Calculer la Trace

### Solution 6.5 : Propri√©t√© $\text{tr}(AB) = \text{tr}(BA)$

```python
# Solution 6.5

A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

# Calculer AB et BA
AB = A @ B
BA = B @ A

# Calculer les traces
trace_AB = np.trace(AB)
trace_BA = np.trace(BA)

print("Matrice AB:")
print(AB)
print(f"tr(AB) = {trace_AB}")
print()

print("Matrice BA:")
print(BA)
print(f"tr(BA) = {trace_BA}")
print()

print("tr(AB) == tr(BA) ?")
print(trace_AB == trace_BA)
print("\nOui ! La propri√©t√© tr(AB) = tr(BA) est v√©rifi√©e. ‚úì")
print("\nRemarque importante:")
print("AB ‚â† BA (matrices diff√©rentes)")
print(np.array_equal(AB, BA))
print("Mais tr(AB) = tr(BA) (traces √©gales) !")
```

### Solution 7.1 : Syst√®me 2√ó2 Simple

```python
# Solution 7.1

# Syst√®me: 2x + y = 5, x + 3y = 6
A = np.array([[2, 1],
              [1, 3]])
b = np.array([5, 6])

# R√©solution
x = np.linalg.solve(A, b)

print("Syst√®me d'√©quations:")
print("2x + y = 5")
print("x + 3y = 6")
print()

print("Forme matricielle Ax = b:")
print("A =")
print(A)
print("b =", b)
print()

print("Solution:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
print()

# V√©rification
print("V√©rification:")
eq1 = 2*x[0] + x[1]
eq2 = x[0] + 3*x[1]
print(f"2√ó{x[0]} + {x[1]} = {eq1} (doit √™tre 5)")
print(f"{x[0]} + 3√ó{x[1]} = {eq2} (doit √™tre 6)")
print("‚úì Solution correcte !")
```

### Solution 7.2 : Syst√®me 3√ó3

```python
# Solution 7.2

# Syst√®me: x + 2y + z = 6, 2x + y - z = 1, x - y + 2z = 5
A = np.array([[1, 2, 1],
              [2, 1, -1],
              [1, -1, 2]])
b = np.array([6, 1, 5])

# R√©solution
x = np.linalg.solve(A, b)

print("Syst√®me d'√©quations:")
print("x + 2y + z = 6")
print("2x + y - z = 1")
print("x - y + 2z = 5")
print()

print("Solution:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
print(f"z = {x[2]}")
print()

# V√©rification
print("V√©rification:")
eq1 = x[0] + 2*x[1] + x[2]
eq2 = 2*x[0] + x[1] - x[2]
eq3 = x[0] - x[1] + 2*x[2]
print(f"√âquation 1: {eq1:.1f} (doit √™tre 6)")
print(f"√âquation 2: {eq2:.1f} (doit √™tre 1)")
print(f"√âquation 3: {eq3:.1f} (doit √™tre 5)")
print("‚úì Solution correcte !")
```
