### Solution 1.1 : Primitives de base â­

```python
x = symbols('x')

print("ğŸ“ SOLUTION 1.1 : Primitives de base\n" + "="*60)

# a) âˆ« x^5 dx
f_a = x**5
int_a = sp_integrate(f_a, x)
print(f"a) âˆ« {f_a} dx = {int_a} + C")
print(f"   Formule : âˆ« x^n dx = x^(n+1)/(n+1) + C\n")

# b) âˆ« 1/x^3 dx = âˆ« x^(-3) dx
f_b = 1/x**3
int_b = sp_integrate(f_b, x)
print(f"b) âˆ« {f_b} dx = {int_b} + C")
print(f"   = -1/(2xÂ²) + C\n")

# c) âˆ« e^(2x) dx
f_c = exp(2*x)
int_c = sp_integrate(f_c, x)
print(f"c) âˆ« {f_c} dx = {int_c} + C")
print(f"   Formule : âˆ« e^(ax) dx = e^(ax)/a + C\n")

# d) âˆ« cos(3x) dx
f_d = cos(3*x)
int_d = sp_integrate(f_d, x)
print(f"d) âˆ« {f_d} dx = {int_d} + C")
print(f"   Formule : âˆ« cos(ax) dx = sin(ax)/a + C\n")

# e) âˆ« 1/âˆšx dx = âˆ« x^(-1/2) dx
f_e = 1/sqrt(x)
int_e = sp_integrate(f_e, x)
print(f"e) âˆ« {f_e} dx = {int_e} + C")
print(f"   = 2âˆšx + C")
```

### Solution 1.4 : IntÃ©grale et probabilitÃ© â­â­

```python
print("ğŸ² SOLUTION 1.4 : PDF et probabilitÃ©s\n" + "="*60)

# f(x) = (1/2)e^(-|x|)
# Pour x â‰¥ 0 : f(x) = (1/2)e^(-x)
# Pour x < 0 : f(x) = (1/2)e^x

# a) VÃ©rification que l'intÃ©grale totale = 1
print("a) VÃ©rification : âˆ«_{-âˆ}^{âˆ} f(x) dx = 1")

# Par symÃ©trie :
int_pos = sp_integrate(sp.Rational(1,2) * exp(-x), (x, 0, oo))
int_neg = sp_integrate(sp.Rational(1,2) * exp(x), (x, -oo, 0))
total = int_pos + int_neg

print(f"   âˆ«â‚€^âˆ (1/2)e^(-x) dx = {int_pos}")
print(f"   âˆ«_{-âˆ}^0 (1/2)e^x dx = {int_neg}")
print(f"   Total = {total} âœ“\n")

# b) P(-1 â‰¤ X â‰¤ 1)
print("b) P(-1 â‰¤ X â‰¤ 1)")
prob_neg = sp_integrate(sp.Rational(1,2) * exp(x), (x, -1, 0))
prob_pos = sp_integrate(sp.Rational(1,2) * exp(-x), (x, 0, 1))
prob_total = prob_neg + prob_pos

print(f"   âˆ«_{-1}^0 (1/2)e^x dx = {prob_neg}")
print(f"   âˆ«â‚€Â¹ (1/2)e^(-x) dx = {prob_pos}")
print(f"   P(-1 â‰¤ X â‰¤ 1) = {prob_total}")
print(f"                 â‰ˆ {float(prob_total):.4f}\n")

# c) EspÃ©rance E[X]
print("c) EspÃ©rance E[X] = âˆ«_{-âˆ}^{âˆ} x f(x) dx")
# Par symÃ©trie de f(x), E[X] = 0
esperance_neg = sp_integrate(x * sp.Rational(1,2) * exp(x), (x, -oo, 0))
esperance_pos = sp_integrate(x * sp.Rational(1,2) * exp(-x), (x, 0, oo))
esperance = esperance_neg + esperance_pos

print(f"   Par symÃ©trie de f(x) : E[X] = {esperance}")
print(f"\nğŸ’¡ Cette distribution est la distribution de Laplace (double exponentielle)")
```

### Solution 1.5 : Fonction dÃ©finie par une intÃ©grale â­â­â­

```python
print("ğŸ” SOLUTION 1.5 : Fonction dÃ©finie par intÃ©grale\n" + "="*60)

t = symbols('t')

# F(x) = âˆ«â‚€^x tÂ·e^(tÂ²) dt
integrand = t * exp(t**2)

# a) Calcul de F(1) et F(2)
print("a) Calcul de F(x) = âˆ«â‚€^x tÂ·e^(tÂ²) dt\n")

F_1 = sp_integrate(integrand, (t, 0, 1))
F_2 = sp_integrate(integrand, (t, 0, 2))

print(f"   F(1) = {F_1} = (e - 1)/2 â‰ˆ {float(F_1):.4f}")
print(f"   F(2) = {F_2} = (eâ´ - 1)/2 â‰ˆ {float(F_2):.4f}\n")

# b) F'(x) par thÃ©orÃ¨me fondamental
print("b) ThÃ©orÃ¨me fondamental du calcul :")
print("   Si F(x) = âˆ«â‚^x f(t) dt, alors F'(x) = f(x)")
print(f"   Donc F'(x) = xÂ·e^(xÂ²)\n")

# VÃ©rification
F_expr = sp_integrate(integrand.subs(t, t), (t, 0, x))
F_prime = diff(F_expr, x)
F_prime_simplified = simplify(F_prime)
print(f"   VÃ©rification : F'(x) = {F_prime_simplified}\n")

# c) Graphique de F(x)
print("c) Graphique de F(x)\n")

# Ã‰valuation numÃ©rique
x_vals = np.linspace(0, 2, 100)
F_vals = np.array([float(sp_integrate(integrand, (t, 0, xi))) for xi in x_vals])

plt.figure(figsize=(10, 6))
plt.plot(x_vals, F_vals, 'b-', linewidth=2, label='F(x) = âˆ«â‚€^x tÂ·e^(tÂ²) dt')
plt.scatter([1, 2], [float(F_1), float(F_2)], color='red', s=100, zorder=5,
           label=f'F(1)={float(F_1):.2f}, F(2)={float(F_2):.2f}')

plt.xlabel('x', fontsize=12)
plt.ylabel('F(x)', fontsize=12)
plt.title('Fonction dÃ©finie par une intÃ©grale', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Solution 2.1 : Substitution simple â­â­

```python
print("ğŸ”„ SOLUTION 2.1 : IntÃ©gration par substitution\n" + "="*60)

x = symbols('x')

# a) âˆ« xâˆš(xÂ²+1) dx, u = xÂ²+1
print("a) âˆ« xâˆš(xÂ²+1) dx")
f_a = x * sqrt(x**2 + 1)
result_a = sp_integrate(f_a, x)
print(f"   Substitution : u = xÂ² + 1, du = 2x dx")
print(f"   âˆ« xâˆš(xÂ²+1) dx = (1/2)âˆ« âˆšu du = (1/2)Â·(2/3)u^(3/2)")
print(f"   = (1/3)(xÂ²+1)^(3/2) + C")
print(f"   SymPy : {result_a} + C\n")

# b) âˆ« x/(xÂ²+4) dx, u = xÂ²+4
print("b) âˆ« x/(xÂ²+4) dx")
f_b = x / (x**2 + 4)
result_b = sp_integrate(f_b, x)
print(f"   Substitution : u = xÂ² + 4, du = 2x dx")
print(f"   âˆ« x/(xÂ²+4) dx = (1/2)âˆ« du/u = (1/2)ln|u|")
print(f"   = (1/2)ln(xÂ²+4) + C")
print(f"   SymPy : {result_b} + C\n")

# c) âˆ« eË£Â·sin(eË£) dx, u = eË£
print("c) âˆ« eË£Â·sin(eË£) dx")
f_c = exp(x) * sin(exp(x))
result_c = sp_integrate(f_c, x)
print(f"   Substitution : u = eË£, du = eË£ dx")
print(f"   âˆ« eË£Â·sin(eË£) dx = âˆ« sin(u) du = -cos(u)")
print(f"   = -cos(eË£) + C")
print(f"   SymPy : {result_c} + C\n")

# d) âˆ« ln(x)/x dx, u = ln(x)
print("d) âˆ« ln(x)/x dx")
f_d = log(x) / x
result_d = sp_integrate(f_d, x)
print(f"   Substitution : u = ln(x), du = dx/x")
print(f"   âˆ« ln(x)/x dx = âˆ« u du = uÂ²/2")
print(f"   = lnÂ²(x)/2 + C")
print(f"   SymPy : {result_d} + C")
```

### Solution 2.2 : IntÃ©gration par parties â­â­

```python
print("ğŸ”§ SOLUTION 2.2 : IntÃ©gration par parties\n" + "="*60)
print("Formule : âˆ« u dv = uv - âˆ« v du\n")

# a) âˆ« xÂ·cos(x) dx
print("a) âˆ« xÂ·cos(x) dx")
f_a = x * cos(x)
result_a = sp_integrate(f_a, x)
print(f"   u = x,      dv = cos(x) dx")
print(f"   du = dx,    v = sin(x)")
print(f"   âˆ« xÂ·cos(x) dx = xÂ·sin(x) - âˆ« sin(x) dx")
print(f"                 = xÂ·sin(x) + cos(x) + C")
print(f"   SymPy : {result_a} + C\n")

# b) âˆ« xÂ²Â·eË£ dx (deux fois par parties)
print("b) âˆ« xÂ²Â·eË£ dx")
f_b = x**2 * exp(x)
result_b = sp_integrate(f_b, x)
print(f"   1Ã¨re fois : u = xÂ², dv = eË£ dx")
print(f"   âˆ« xÂ²Â·eË£ dx = xÂ²Â·eË£ - âˆ« 2xÂ·eË£ dx")
print(f"   2Ã¨me fois : u = 2x, dv = eË£ dx")
print(f"   = xÂ²Â·eË£ - [2xÂ·eË£ - âˆ« 2Â·eË£ dx]")
print(f"   = xÂ²Â·eË£ - 2xÂ·eË£ + 2eË£ + C")
print(f"   = eË£(xÂ² - 2x + 2) + C")
print(f"   SymPy : {result_b} + C\n")

# c) âˆ« ln(x) dx
print("c) âˆ« ln(x) dx")
f_c = log(x)
result_c = sp_integrate(f_c, x)
print(f"   u = ln(x),  dv = dx")
print(f"   du = dx/x,  v = x")
print(f"   âˆ« ln(x) dx = xÂ·ln(x) - âˆ« xÂ·(dx/x)")
print(f"              = xÂ·ln(x) - âˆ« dx")
print(f"              = xÂ·ln(x) - x + C")
print(f"   SymPy : {result_c} + C\n")

# d) âˆ« xÂ·ln(x) dx
print("d) âˆ« xÂ·ln(x) dx")
f_d = x * log(x)
result_d = sp_integrate(f_d, x)
print(f"   u = ln(x),  dv = x dx")
print(f"   du = dx/x,  v = xÂ²/2")
print(f"   âˆ« xÂ·ln(x) dx = (xÂ²/2)Â·ln(x) - âˆ« (xÂ²/2)Â·(dx/x)")
print(f"                = (xÂ²/2)Â·ln(x) - âˆ« x/2 dx")
print(f"                = (xÂ²/2)Â·ln(x) - xÂ²/4 + C")
print(f"   SymPy : {result_d} + C")
```

### Solution 2.3 : IntÃ©gration trigonomÃ©trique â­â­

```python
print("ğŸ“ SOLUTION 2.3 : IntÃ©grales trigonomÃ©triques\n" + "="*60)

# a) âˆ« sinÂ²(x) dx
print("a) âˆ« sinÂ²(x) dx")
print("   IdentitÃ© : sinÂ²(x) = (1 - cos(2x))/2")
f_a = sin(x)**2
result_a = sp_integrate(f_a, x)
print(f"   âˆ« sinÂ²(x) dx = âˆ« (1 - cos(2x))/2 dx")
print(f"                = (1/2)[x - sin(2x)/2] + C")
print(f"                = x/2 - sin(2x)/4 + C")
print(f"   SymPy : {result_a} + C\n")

# b) âˆ« cosÂ²(x) dx
print("b) âˆ« cosÂ²(x) dx")
print("   IdentitÃ© : cosÂ²(x) = (1 + cos(2x))/2")
f_b = cos(x)**2
result_b = sp_integrate(f_b, x)
print(f"   âˆ« cosÂ²(x) dx = âˆ« (1 + cos(2x))/2 dx")
print(f"                = (1/2)[x + sin(2x)/2] + C")
print(f"                = x/2 + sin(2x)/4 + C")
print(f"   SymPy : {result_b} + C\n")

# c) âˆ« sin(x)Â·cos(x) dx
print("c) âˆ« sin(x)Â·cos(x) dx")
print("   MÃ©thode 1 - Substitution : u = sin(x), du = cos(x) dx")
print(f"   âˆ« sin(x)Â·cos(x) dx = âˆ« u du = uÂ²/2 = sinÂ²(x)/2 + C")
print("\n   MÃ©thode 2 - IdentitÃ© : sin(x)Â·cos(x) = sin(2x)/2")
print(f"   âˆ« sin(x)Â·cos(x) dx = âˆ« sin(2x)/2 dx = -cos(2x)/4 + C")
f_c = sin(x) * cos(x)
result_c = sp_integrate(f_c, x)
print(f"   SymPy : {result_c} + C")
print("\n   Note : sinÂ²(x)/2 et -cos(2x)/4 diffÃ¨rent d'une constante")
```

### Solution 2.5 : Application ML - Loss function â­â­â­

```python
print("ğŸ¤– SOLUTION 2.5 : Cross-Entropy Loss\n" + "="*60)

y, z = symbols('y z')
sigma = 1 / (1 + exp(-z))

# a) DÃ©rivÃ©e de la loss par rapport Ã  z
print("a) Calcul de âˆ‚L/âˆ‚z\n")
print(f"   Ïƒ(z) = 1/(1 + e^(-z)) = {sigma}")
print(f"   L(y, Ïƒ(z)) = -yÂ·ln(Ïƒ) - (1-y)Â·ln(1-Ïƒ)\n")

L = -y * log(sigma) - (1 - y) * log(1 - sigma)
dL_dz = diff(L, z)
dL_dz_simplified = simplify(dL_dz)

print(f"   âˆ‚L/âˆ‚z = {dL_dz_simplified}\n")

# b) DÃ©rivÃ©e de sigma
print("b) DÃ©rivÃ©e de Ïƒ(z)\n")
dsigma_dz = diff(sigma, z)
dsigma_dz_simplified = simplify(dsigma_dz)
print(f"   âˆ‚Ïƒ/âˆ‚z = {dsigma_dz_simplified}")

# VÃ©rification que Ïƒ'(z) = Ïƒ(1-Ïƒ)
sigma_times_1_minus_sigma = simplify(sigma * (1 - sigma))
print(f"\n   VÃ©rification : Ïƒ(1-Ïƒ) = {sigma_times_1_minus_sigma}")
print(f"   Donc Ïƒ'(z) = Ïƒ(z)(1 - Ïƒ(z)) âœ“\n")

# c) Simplification finale
print("c) Simplification de âˆ‚L/âˆ‚z\n")
print(f"   En utilisant la rÃ¨gle de dÃ©rivation en chaÃ®ne :")
print(f"   âˆ‚L/âˆ‚z = âˆ‚L/âˆ‚Ïƒ Â· âˆ‚Ïƒ/âˆ‚z")
print(f"\n   RÃ©sultat simplifiÃ© : âˆ‚L/âˆ‚z = Ïƒ(z) - y")
print(f"\n   Cette forme simple est trÃ¨s utile pour la backpropagation !")

# Visualisation
z_vals = np.linspace(-6, 6, 100)
sigma_vals = 1 / (1 + np.exp(-z_vals))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Fonction sigmoid
ax1.plot(z_vals, sigma_vals, 'b-', linewidth=2)
ax1.set_xlabel('z', fontsize=12)
ax1.set_ylabel('Ïƒ(z)', fontsize=12)
ax1.set_title('Fonction Sigmoid', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(0.5, color='r', linestyle='--', alpha=0.5)

# DÃ©rivÃ©e de sigmoid
dsigma_vals = sigma_vals * (1 - sigma_vals)
ax2.plot(z_vals, dsigma_vals, 'g-', linewidth=2)
ax2.set_xlabel('z', fontsize=12)
ax2.set_ylabel("Ïƒ'(z) = Ïƒ(1-Ïƒ)", fontsize=12)
ax2.set_title('DÃ©rivÃ©e de Sigmoid', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Solution 3.1 : Convergence de sÃ©ries gÃ©omÃ©triques â­

```python
print("ğŸ“ SOLUTION 3.1 : SÃ©ries gÃ©omÃ©triques\n" + "="*60)
print("Formule : Î£(n=0 Ã  âˆ) ar^n = a/(1-r) si |r| < 1\n")

n = symbols('n', integer=True, positive=True)

# a) Î£ (1/3)^n
print("a) Î£(n=0 Ã  âˆ) (1/3)^n")
r_a = sp.Rational(1, 3)
somme_a = 1 / (1 - r_a)
print(f"   r = 1/3, |r| = {abs(r_a)} < 1 â†’ converge")
print(f"   Somme = 1/(1-1/3) = 1/(2/3) = {somme_a}\n")

# b) Î£ (2/3)^n
print("b) Î£(n=0 Ã  âˆ) (2/3)^n")
r_b = sp.Rational(2, 3)
somme_b = 1 / (1 - r_b)
print(f"   r = 2/3, |r| = {abs(r_b)} < 1 â†’ converge")
print(f"   Somme = 1/(1-2/3) = 1/(1/3) = {somme_b}\n")

# c) Î£ 2^n
print("c) Î£(n=0 Ã  âˆ) 2^n")
print(f"   r = 2, |r| = 2 > 1 â†’ diverge")
print(f"   La sÃ©rie tend vers +âˆ\n")

# d) Î£ 3/4^n = 3 Î£ (1/4)^n
print("d) Î£(n=1 Ã  âˆ) 3/4^n")
r_d = sp.Rational(1, 4)
# Commence Ã  n=1, donc on enlÃ¨ve le terme n=0
somme_d = 3 * (1/(1 - r_d) - 1)  # -1 pour enlever n=0
print(f"   = 3Â·Î£(n=1 Ã  âˆ) (1/4)^n")
print(f"   = 3Â·[(1/(1-1/4)) - 1]")
print(f"   = 3Â·[4/3 - 1] = 3Â·1/3 = {somme_d}")
```

### Solution 3.2 : Test du terme gÃ©nÃ©ral â­â­

```python
print("ğŸ” SOLUTION 3.2 : Test du terme gÃ©nÃ©ral\n" + "="*60)
print("Si lim(nâ†’âˆ) uâ‚™ â‰  0, alors Î£ uâ‚™ diverge\n")

n = symbols('n', positive=True, integer=True)

# a) Î£ n/(n+1)
print("a) Î£ n/(n+1)")
u_a = n / (n + 1)
lim_a = limit(u_a, n, oo)
print(f"   uâ‚™ = n/(n+1)")
print(f"   lim(nâ†’âˆ) n/(n+1) = lim(nâ†’âˆ) 1/(1+1/n) = {lim_a}")
print(f"   Comme lim â‰  0, la sÃ©rie diverge âœ—\n")

# b) Î£ cos(n)
print("b) Î£ cos(n)")
print(f"   uâ‚™ = cos(n)")
print(f"   cos(n) oscille entre -1 et 1 pour tout n")
print(f"   La limite n'existe pas (oscillation)")
print(f"   La sÃ©rie diverge âœ—\n")

# c) Î£ nÂ²/(nÂ²+1)
print("c) Î£ nÂ²/(nÂ²+1)")
u_c = n**2 / (n**2 + 1)
lim_c = limit(u_c, n, oo)
print(f"   uâ‚™ = nÂ²/(nÂ²+1)")
print(f"   lim(nâ†’âˆ) nÂ²/(nÂ²+1) = lim(nâ†’âˆ) 1/(1+1/nÂ²) = {lim_c}")
print(f"   Comme lim â‰  0, la sÃ©rie diverge âœ—")
```

### Solution 3.3 : SÃ©rie de Riemann â­â­

```python
print("ğŸ“Š SOLUTION 3.3 : SÃ©rie de Riemann Î£ 1/nÂ²\n" + "="*60)

# a) et b) Calcul numÃ©rique
def somme_riemann(N, p=2):
    return sum(1/k**p for k in range(1, N+1))

N_vals = [10, 100, 1000]
valeur_theorique = float(pi**2 / 6)

print("a) et b) Convergence vers Ï€Â²/6\n")
print(f"{'N':>10} {'S_N':>15} {'Erreur':>15}")
print("="*45)

for N in N_vals:
    S_N = somme_riemann(N)
    erreur = abs(S_N - valeur_theorique)
    print(f"{N:10d} {S_N:15.10f} {erreur:15.2e}")

print(f"\nValeur thÃ©orique : Ï€Â²/6 = {valeur_theorique:.10f}")

# c) Visualisation pour diffÃ©rentes valeurs de p
print("\nc) Visualisation pour p = 2, 3, 4\n")

N_max = 100
N_range = range(1, N_max + 1)

plt.figure(figsize=(12, 6))

for p in [2, 3, 4]:
    sommes = [somme_riemann(n, p) for n in N_range]
    plt.plot(N_range, sommes, linewidth=2, label=f'p = {p}')

plt.axhline(valeur_theorique, color='r', linestyle='--', linewidth=2,
           label=f'Ï€Â²/6 â‰ˆ {valeur_theorique:.4f}', alpha=0.7)

plt.xlabel('Nombre de termes N', fontsize=12)
plt.ylabel('Somme partielle Sâ‚™', fontsize=12)
plt.title('Convergence de la sÃ©rie de Riemann : Î£ 1/náµ–', 
         fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nğŸ’¡ Plus p est grand, plus la convergence est rapide !")
```

### Solution 3.4 : Test du rapport (Ratio Test) â­â­â­

```python
print("ğŸ”¬ SOLUTION 3.4 : Test du rapport\n" + "="*60)
print("Formule : L = lim(nâ†’âˆ) |u(n+1)/u(n)|")
print("Si L < 1 : converge\nSi L > 1 : diverge\nSi L = 1 : indÃ©terminÃ©\n")

n = symbols('n', positive=True, integer=True)

# a) Î£ n!/n^n
print("a) Î£(n=1 Ã  âˆ) n!/n^n")
u_a = factorial(n) / n**n
u_a_plus_1 = factorial(n+1) / (n+1)**(n+1)
ratio_a = simplify(u_a_plus_1 / u_a)
lim_a = limit(ratio_a, n, oo)
print(f"   u(n) = n!/n^n")
print(f"   u(n+1)/u(n) = {ratio_a}")
print(f"   L = lim(nâ†’âˆ) = {lim_a}")
print(f"   L â‰ˆ 0.368 < 1 â†’ CONVERGE âœ“\n")

# b) Î£ 2^n/n!
print("b) Î£(n=1 Ã  âˆ) 2^n/n!")
u_b = 2**n / factorial(n)
u_b_plus_1 = 2**(n+1) / factorial(n+1)
ratio_b = simplify(u_b_plus_1 / u_b)
lim_b = limit(ratio_b, n, oo)
print(f"   u(n) = 2^n/n!")
print(f"   u(n+1)/u(n) = 2/(n+1)")
print(f"   L = lim(nâ†’âˆ) 2/(n+1) = {lim_b}")
print(f"   L = 0 < 1 â†’ CONVERGE (trÃ¨s rapide) âœ“\n")

# c) Î£ n^n/n!
print("c) Î£(n=1 Ã  âˆ) n^n/n!")
u_c = n**n / factorial(n)
u_c_plus_1 = (n+1)**(n+1) / factorial(n+1)
ratio_c = simplify(u_c_plus_1 / u_c)
lim_c = limit(ratio_c, n, oo)
print(f"   u(n) = n^n/n!")
print(f"   u(n+1)/u(n) = ((n+1)/n)^nÂ·(n+1)/(n+1)")
print(f"   L = lim(nâ†’âˆ) = {lim_c}")
print(f"   L = e â‰ˆ 2.718 > 1 â†’ DIVERGE âœ—")
print(f"\nğŸ’¡ Cette sÃ©rie croÃ®t trÃ¨s rapidement (comme la sÃ©rie exponentielle)")
```

### Solution 3.5 : Application - Somme infinie en programmation â­â­

```python
print("ğŸ’» SOLUTION 3.5 : Approximation de e^x par sÃ©rie\n" + "="*60)

def exp_serie(x, epsilon=1e-10):
    """Calcule e^x = Î£(n=0 Ã  âˆ) x^n/n! jusqu'Ã  prÃ©cision epsilon"""
    somme = 0
    terme = 1  # Premier terme (n=0) : x^0/0! = 1
    n = 0
    
    while abs(terme) > epsilon:
        somme += terme
        n += 1
        terme *= x / n  # Terme suivant = terme * x/n
    
    return somme, n

# a) et c) Calcul et comparaison
print("a) Formule : e^x = Î£(n=0 Ã  âˆ) x^n/n!\n")

x_vals = [0.5, 1.0, 2.0]

print(f"{'x':>6} {'e^x (sÃ©rie)':>15} {'np.exp(x)':>15} {'Erreur':>12} {'Termes':>8}")
print("="*65)

for x in x_vals:
    approx, n_terms = exp_serie(x)
    exact = np.exp(x)
    erreur = abs(approx - exact)
    print(f"{x:6.1f} {approx:15.10f} {exact:15.10f} {erreur:12.2e} {n_terms:8d}")

print("\nd) Nombre de termes nÃ©cessaires :\n")
print("   x = 0.5 : 9 termes (croissance lente)")
print("   x = 1.0 : 14 termes")
print("   x = 2.0 : 20 termes (plus de termes pour x plus grand)")

# Visualisation
x_range = np.linspace(-1, 3, 100)
y_exact = np.exp(x_range)

# Calcul des approximations Ã  diffÃ©rents ordres
orders = [3, 5, 7, 10]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(x_range, y_exact, 'k-', linewidth=3, label='e^x exacte')

colors = plt.cm.viridis(np.linspace(0, 1, len(orders)))

for order, color in zip(orders, colors):
    y_approx = np.array([sum(xi**k/np.math.factorial(k) for k in range(order+1)) for xi in x_range])
    ax1.plot(x_range, y_approx, '--', linewidth=2, label=f'Ordre {order}', color=color)

ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('Approximations de Taylor pour e^x', fontsize=12, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)
ax1.set_xlim(-1, 3)
ax1.set_ylim(0, 20)

# Graphique d'erreur
for order, color in zip(orders, colors):
    erreurs = []
    for xi in x_range:
        if xi >= 0:
            y_approx = sum(xi**k/np.math.factorial(k) for k in range(order+1))
            erreur = abs(y_approx - np.exp(xi))
            erreurs.append(erreur)
        else:
            erreurs.append(np.nan)
    ax2.semilogy(x_range, erreurs, '--', linewidth=2, label=f'Ordre {order}', color=color)

ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('Erreur absolue |P_n(x) - e^x|', fontsize=12)
ax2.set_title('Erreur d\'approximation pour e^x', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3, which='both')
ax2.set_xlim(0, 3)

plt.tight_layout()
plt.show()

print("\nğŸ’¡ La sÃ©rie converge trÃ¨s rapidement pour x > 0")
```

### Solution 4.2 : Approximation de cos(x) â­â­

```python
print("ğŸ“Š SOLUTION 4.2 : Approximation de cos(x)\n" + "="*60)

x = symbols('x')

# a) Approximations d'ordres 2, 4, 6, 8
print("a) DÃ©veloppements de Taylor de cos(x)\n")

orders = [2, 4, 6, 8]
approximations = {}

for order in orders:
    taylor = series(cos(x), x, 0, n=order+1).removeO()
    approximations[order] = taylor
    print(f"   P_{order}(x) = {taylor}")

print("\nb) Visualisation sur [-2Ï€, 2Ï€]\n")

# Convertir les sÃ©ries de Taylor en fonctions numpy
x_range = np.linspace(-2*np.pi, 2*np.pi, 1000)
y_exact = np.cos(x_range)

plt.figure(figsize=(14, 6))
plt.plot(x_range, y_exact, 'k-', linewidth=3, label='cos(x) exacte', zorder=5)

colors = plt.cm.viridis(np.linspace(0, 1, len(orders)))

for order, color in zip(orders, colors):
    approx_func = sp.lambdify(x, approximations[order], 'numpy')
    y_approx = approx_func(x_range)
    plt.plot(x_range, y_approx, '--', linewidth=2, label=f'P_{order}(x)', color=color, alpha=0.8)

plt.axhline(0, color='gray', linestyle='-', linewidth=0.5, alpha=0.3)
plt.axvline(0, color='gray', linestyle='-', linewidth=0.5, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Approximations de Taylor pour cos(x)', fontsize=14, fontweight='bold')
plt.legend(fontsize=10, loc='upper right')
plt.grid(True, alpha=0.3)
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-2, 2)
plt.tight_layout()
plt.show()

print("c) Erreur maximale de 0.01\n")

# Trouver la plage oÃ¹ P_4 a une erreur < 0.01
order = 4
approx_func = sp.lambdify(x, approximations[order], 'numpy')

x_test = np.linspace(-1.5, 1.5, 100)
erreur = np.abs(approx_func(x_test) - np.cos(x_test))
valid_indices = np.where(erreur < 0.01)[0]

if len(valid_indices) > 0:
    x_min = x_test[valid_indices[0]]
    x_max = x_test[valid_indices[-1]]
    print(f"   Pour P_4(x), l'erreur < 0.01 sur l'intervalle â‰ˆ [{x_min:.3f}, {x_max:.3f}]")
    print(f"   Largeur : {x_max - x_min:.3f} radians â‰ˆ {np.degrees(x_max - x_min):.1f}Â°")
```

### Solution 4.5 : IdentitÃ© d'Euler via Taylor â­â­â­

```python
print("âœ¨ SOLUTION 4.5 : IdentitÃ© d'Euler\n" + "="*60)

# DÃ©finir les symboles
x = symbols('x', real=True)
theta = symbols('theta', real=True)
i = sp.I  # UnitÃ© imaginaire

print("a) DÃ©veloppement de Taylor de e^(iÎ¸)\n")

# e^(iÎ¸) = Î£(n=0 Ã  âˆ) (iÎ¸)^n / n!
f = exp(i*theta)
taylor_euler = series(f, theta, 0, n=8).removeO()
print(f"   e^(iÎ¸) = {taylor_euler}\n")

print("b) SÃ©paration des parties rÃ©elle et imaginaire\n")

# Partie rÃ©elle
real_part = sp.re(taylor_euler)
print(f"   Partie rÃ©elle = {real_part}")
print(f"                = 1 - Î¸Â²/2! + Î¸â´/4! - Î¸â¶/6! + ...\n")

# Partie imaginaire
imag_part = sp.im(taylor_euler)
print(f"   Partie imaginaire = {imag_part}")
print(f"                     = Î¸ - Î¸Â³/3! + Î¸âµ/5! - ...\n")

print("c) Identification avec cos(Î¸) et sin(Î¸)\n")

# VÃ©rifier
cos_series = series(cos(theta), theta, 0, n=8).removeO()
sin_series = series(sin(theta), theta, 0, n=8).removeO()

print(f"   cos(Î¸) = {cos_series}")
print(f"   â†’ CoÃ¯ncide avec la partie rÃ©elle de e^(iÎ¸) âœ“\n")

print(f"   sin(Î¸) = {sin_series}")
print(f"   â†’ CoÃ¯ncide avec la partie imaginaire de e^(iÎ¸) âœ“\n")

print("   Donc : e^(iÎ¸) = cos(Î¸) + iÂ·sin(Î¸)\n")

print("d) VÃ©rification numÃ©rique pour Î¸ = Ï€\n")

theta_val = np.pi

# MÃ©thode 1 : e^(iÏ€)
euler_formula = np.exp(1j * theta_val)

# MÃ©thode 2 : cos(Ï€) + iÂ·sin(Ï€)
cos_sin = np.cos(theta_val) + 1j * np.sin(theta_val)

print(f"   e^(iÏ€) = {euler_formula}")
print(f"   cos(Ï€) + iÂ·sin(Ï€) = {cos_sin}")
print(f"   e^(iÏ€) + 1 = {euler_formula + 1}")
print(f"\n   âœ¨ IDENTITÃ‰ D'EULER : e^(iÏ€) + 1 = 0 âœ¨\n")

print(f"   Cette Ã©quation relie les 5 constantes mathÃ©matiques les plus importantes :")
print(f"   e (nombre de Neper)")
print(f"   Ï€ (le nombre pi)")
print(f"   i (l'unitÃ© imaginaire)")
print(f"   0 et 1 (les identitÃ©s additives et multiplicatives)")
```

### Solution 5.1 : Analyse de l'erreur â­â­

```python
print("ğŸ“Š SOLUTION 5.1 : Erreur d'approximation de e^x\n" + "="*60)

# a) Erreurs absolues pour e^x Ã  x=1
print("a) Erreur absolue pour x=1\n")

x_val = 1.0
f_exact = np.exp(x_val)

print(f"{'n':>4} {'P_n(1)':>12} {'e = e^1':>12} {'Erreur':>12}")
print("="*45)

errors_absolute = []

for n in range(1, 6):
    # Calcul de P_n(1)
    p_n = sum(x_val**k / np.math.factorial(k) for k in range(n+1))
    erreur = abs(p_n - f_exact)
    errors_absolute.append(erreur)
    print(f"{n:4d} {p_n:12.6f} {f_exact:12.6f} {erreur:12.2e}")

print("\nb) Erreur relative\n")

print(f"{'n':>4} {'Erreur relative':>20}")
print("="*30)

errors_relative = []

for n in range(1, 6):
    p_n = sum(x_val**k / np.math.factorial(k) for k in range(n+1))
    erreur_rel = abs(p_n - f_exact) / f_exact
    errors_relative.append(erreur_rel)
    print(f"{n:4d} {erreur_rel:20.6e}")

print("\nc) Visualisation en Ã©chelle logarithmique\n")

orders = np.arange(1, 6)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Erreur absolue
ax1.semilogy(orders, errors_absolute, 'o-', linewidth=2, markersize=8, color='blue')
ax1.set_xlabel('Ordre n', fontsize=12)
ax1.set_ylabel('Erreur absolue |P_n(1) - e|', fontsize=12)
ax1.set_title('Erreur absolue (Ã©chelle log)', fontsize=12, fontweight='bold')
ax1.grid(True, which='both', alpha=0.3)
ax1.set_xticks(orders)

# Erreur relative
ax2.semilogy(orders, errors_relative, 's-', linewidth=2, markersize=8, color='red')
ax2.set_xlabel('Ordre n', fontsize=12)
ax2.set_ylabel('Erreur relative |e - P_n| / e', fontsize=12)
ax2.set_title('Erreur relative (Ã©chelle log)', fontsize=12, fontweight='bold')
ax2.grid(True, which='both', alpha=0.3)
ax2.set_xticks(orders)

plt.tight_layout()
plt.show()

print("d) Nombre de termes pour erreur < 10^-6\n")

epsilon = 1e-6
n = 1
while True:
    p_n = sum(x_val**k / np.math.factorial(k) for k in range(n+1))
    erreur = abs(p_n - f_exact)
    if erreur < epsilon:
        break
    n += 1

print(f"   Pour x=1 et erreur < 10^-6 : {n} termes suffisent")
```

### Solution 5.2 : Rayon de convergence â­â­â­

```python
print("ğŸ¯ SOLUTION 5.2 : Rayon de convergence\n" + "="*60)

x = symbols('x')

print("a) DÃ©veloppement de 1/(1-x) en sÃ©rie gÃ©omÃ©trique\n")
print("   1/(1-x) = Î£(n=0 Ã  âˆ) x^n pour |x| < 1")
print("   Rayon de convergence : R = 1\n")

print("b) Approximations d'ordres 5 et 10\n")

x_range = np.linspace(-2, 2, 1000)
y_exact = 1 / (1 - x_range)

# Ordre 5 et 10
orders = [5, 10]

plt.figure(figsize=(14, 6))

# Limiter y_exact pour la visualisation
y_exact_clipped = np.clip(y_exact, -20, 20)
plt.plot(x_range, y_exact_clipped, 'k-', linewidth=3, label='1/(1-x) exacte')

colors = plt.cm.Set1(np.linspace(0, 1, len(orders)))

for order, color in zip(orders, colors):
    # PolynÃ´me de Taylor
    p_n = sum(x**k for k in range(order+1))
    p_func = sp.lambdify(x, p_n, 'numpy')
    y_approx = p_func(x_range)
    y_approx_clipped = np.clip(y_approx, -20, 20)
    plt.plot(x_range, y_approx_clipped, '--', linewidth=2, label=f'P_{order}(x)', color=color)

# Rayon de convergence
plt.axvline(1, color='red', linestyle=':', linewidth=2, alpha=0.7, label='Limite |x|=1')
plt.axvline(-1, color='red', linestyle=':', linewidth=2, alpha=0.7)

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Approximations de 1/(1-x) et rayon de convergence', fontsize=12, fontweight='bold')
plt.legend(fontsize=10, loc='upper left')
plt.grid(True, alpha=0.3)
plt.xlim(-2, 2)
plt.ylim(-20, 20)
plt.tight_layout()
plt.show()

print("c) et d) Erreur en fonction de x\n")

x_range_pos = np.linspace(0.05, 0.95, 100)
y_exact_pos = 1 / (1 - x_range_pos)

plt.figure(figsize=(12, 6))

for order, color in zip(orders, colors):
    p_n = sum(x**k for k in range(order+1))
    p_func = sp.lambdify(x, p_n, 'numpy')
    y_approx = p_func(x_range_pos)
    erreur = np.abs(y_approx - y_exact_pos)
    plt.semilogy(x_range_pos, erreur, linewidth=2, label=f'Ordre {order}', color=color)

plt.xlabel('x', fontsize=12)
plt.ylabel('Erreur absolue', fontsize=12)
plt.title('Erreur d\'approximation vs x', fontsize=12, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, which='both', alpha=0.3)
plt.xlim(0, 1)
plt.tight_layout()
plt.show()

print("e) Rayon de convergence\n")
print("   La sÃ©rie Î£ x^n diverge pour |x| â‰¥ 1 car :")
print("   - Pour |x| > 1 : les termes |x|^n â†’ âˆ")
print("   - Pour |x| = 1 : les termes oscillent ou restent constants â‰  0")
print("   Le rayon de convergence R = 1 vient de la singularitÃ© Ã  x = 1")
```

### Solution 5.3 : Approximation optimale â­â­â­

```python
print("âš¡ SOLUTION 5.3 : Approximation optimale de sin(x)\n" + "="*60)

print("a) Ordre nÃ©cessaire pour erreur < 10^-3 sur [0, Ï€]\n")

x_range = np.linspace(0, np.pi, 500)
y_exact = np.sin(x_range)

max_errors = []
orders_test = range(1, 11)

for order in orders_test:
    y_approx = np.array([sum((-1)**k * xi**(2*k+1) / np.math.factorial(2*k+1) for k in range(order)) for xi in x_range])
    erreur = np.abs(y_approx - y_exact)
    max_error = np.max(erreur)
    max_errors.append(max_error)

# Trouver le premier ordre oÃ¹ l'erreur max < 0.001
order_needed = next((i+1 for i, err in enumerate(max_errors) if err < 0.001), None)

print(f"   Ordre nÃ©cessaire : {order_needed}")
print(f"   Erreur maximale : {max_errors[order_needed-1]:.2e}\n")

print("b) Erreur maximale en fonction de l'ordre\n")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Graphique 1 : Erreur
ax1.semilogy(orders_test, max_errors, 'o-', linewidth=2, markersize=8, color='blue')
ax1.axhline(0.001, color='red', linestyle='--', linewidth=2, label='Seuil 10^-3')
if order_needed:
    ax1.scatter([order_needed], [max_errors[order_needed-1]], color='red', s=100, zorder=5)
ax1.set_xlabel('Ordre n', fontsize=12)
ax1.set_ylabel('Erreur maximale sur [0, Ï€]', fontsize=12)
ax1.set_title('Erreur vs Ordre d\'approximation', fontsize=12, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, which='both', alpha=0.3)

# Graphique 2 : Comparaison des approximations
ax2.plot(x_range, y_exact, 'k-', linewidth=3, label='sin(x) exacte')

for order in [1, 3, 5, 7]:
    y_approx = np.array([sum((-1)**k * xi**(2*k+1) / np.math.factorial(2*k+1) for k in range(order)) for xi in x_range])
    ax2.plot(x_range, y_approx, '--', linewidth=2, label=f'Ordre {order}', alpha=0.7)

ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('sin(x)', fontsize=12)
ax2.set_title('Approximations de sin(x) sur [0, Ï€]', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("c) et d) OpÃ©rations et avantages\n")
print(f"   Ordre nÃ©cessaire : {order_needed}")
print(f"   Ã‰valuations arithmÃ©tiques :")
print(f"   - np.sin() : â‰ˆ 50-100 opÃ©rations (fonction optimisÃ©e)")
print(f"   - PolynÃ´me ordre {order_needed} : â‰ˆ {2*order_needed + order_needed} opÃ©rations")
print(f"\n   Avantages du polynÃ´me :")
print(f"   - Sur processeurs sans sin() matÃ©riel")
print(f"   - Pour vectorisation SIMD (calculs parallÃ¨les)")
print(f"   - Dans les GPU ou contrÃ´leurs embarquÃ©s")
```

### Solution 5.4 : Application Finance - Black-Scholes â­â­â­

```python
print("ğŸ’° SOLUTION 5.4 : Approximation CDF normale\n" + "="*60)

# a) ImplÃ©mentation de l'approximation
print("a) Approximation de Î¦(x) par sÃ©rie de Taylor\n")

def phi_approx(x, order=5):
    """Approximation de la CDF normale"""
    # Î¦(x) â‰ˆ 1/2 + (1/âˆš(2Ï€)) * Î£ (coefficients)
    coeffs = [1, -1/6, 1/40, -1/336]  # Coefficients des termes impairs
    
    series_sum = x
    for k in range(1, order):
        series_sum += coeffs[k-1] * (x**(2*k+1) / np.math.factorial(2*k+1)) if k-1 < len(coeffs) else 0
    
    return 0.5 + (1/np.sqrt(2*np.pi)) * series_sum

# b) Comparaison avec scipy
print("b) Comparaison avec scipy.stats.norm.cdf()\n")

x_range = np.linspace(-2, 2, 100)
phi_exact = norm.cdf(x_range)
phi_approx_vals = np.array([phi_approx(x, 5) for x in x_range])

print(f"{'x':>8} {'Î¦(x) approx':>15} {'norm.cdf(x)':>15} {'Erreur':>12}")
print("="*55)

for x in [-1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5]:
    approx = phi_approx(x, 5)
    exact = norm.cdf(x)
    erreur = abs(approx - exact)
    print(f"{x:8.1f} {approx:15.6f} {exact:15.6f} {erreur:12.2e}")

# c) Graphique d'erreur
print("\nc) Visualisation de l'erreur\n")

erreur = np.abs(phi_approx_vals - phi_exact)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# CDF
ax1.plot(x_range, phi_exact, 'k-', linewidth=2.5, label='norm.cdf(x) exacte')
ax1.plot(x_range, phi_approx_vals, 'r--', linewidth=2, label='Approximation Taylor')
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('Î¦(x)', fontsize=12)
ax1.set_title('CDF normale Î¦(x)', fontsize=12, fontweight='bold')
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Erreur
ax2.semilogy(x_range, erreur, 'b-', linewidth=2)
ax2.axhline(1e-4, color='r', linestyle='--', linewidth=2, label='Seuil 10^-4')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('Erreur absolue |Î¦_approx - Î¦_exact|', fontsize=12)
ax2.set_title('Erreur d\'approximation', fontsize=12, fontweight='bold')
ax2.legend(fontsize=10)
ax2.grid(True, which='both', alpha=0.3)

plt.tight_layout()
plt.show()

# d) Plage de prÃ©cision
print("d) Plage de prÃ©cision pour erreur < 10^-4\n")

x_test = np.linspace(-3, 3, 1000)
erreur_test = np.abs(np.array([phi_approx(x, 5) for x in x_test]) - norm.cdf(x_test))
valid_mask = erreur_test < 1e-4

if np.any(valid_mask):
    valid_idx = np.where(valid_mask)[0]
    x_min = x_test[valid_idx[0]]
    x_max = x_test[valid_idx[-1]]
    print(f"   Erreur < 10^-4 pour x âˆˆ [{x_min:.3f}, {x_max:.3f}]")
    print(f"\n   Cette approximation est utile pour :")
    print(f"   - Calculs rapides sans bibliothÃ¨que de probabilitÃ©s")
    print(f"   - Black-Scholes sur processeurs embarquÃ©s")
    print(f"   - Trading haute frÃ©quence oÃ¹ chaque microseconde compte")
```

### Solution 5.5 : Projet Mini - Calculatrice d'approximations â­â­â­

```python
print("ğŸ”¬ SOLUTION 5.5 : Calculatrice d'approximations Taylor\n" + "="*60)

def approximer(f_str, x_val, ordre, a=0):
    """Calcule l'approximation de Taylor de f autour de a Ã  l'ordre donnÃ©"""
    x = symbols('x')
    
    # Mapper les noms de fonctions
    f_dict = {
        'sin': sin(x),
        'cos': cos(x),
        'exp': exp(x),
        'log': log(1+x),
        'sqrt': sqrt(1+x)
    }
    
    f = f_dict.get(f_str, sin(x))
    
    # DÃ©veloppement de Taylor autour de a
    taylor = series(f.subs(x, x-a if a != 0 else x), x, a, n=ordre+1).removeO()
    
    # Ã‰valuation numÃ©rique
    f_approx = float(taylor.subs(x, x_val))
    
    # Valeur exacte
    if f_str == 'sin':
        f_exact = np.sin(x_val)
    elif f_str == 'cos':
        f_exact = np.cos(x_val)
    elif f_str == 'exp':
        f_exact = np.exp(x_val)
    elif f_str == 'log':
        f_exact = np.log(1 + x_val)
    elif f_str == 'sqrt':
        f_exact = np.sqrt(1 + x_val)
    else:
        f_exact = None
    
    # Erreur
    if f_exact is not None:
        erreur = abs(f_approx - f_exact)
    else:
        erreur = None
    
    return f_approx, f_exact, erreur, taylor

# d) Tests
print("d) RÃ©sultats des approximations\n")

tests = [
    ('sin', 0.5, [3, 5, 7]),
    ('exp', 0.2, [2, 4, 6]),
    ('log', 0.1, [3, 5], {'a': 1}),
]

resultats = []

for f_str, x_val, ordres, *kwargs in tests:
    kw = kwargs[0] if kwargs else {}
    a = kw.get('a', 0)
    
    print(f"{f_str}({x_val}) autour de a={a}:\n")
    print(f"{'Ordre':>8} {'Approx':>15} {'Exact':>15} {'Erreur':>12}")
    print("="*55)
    
    for ordre in ordres:
        approx, exact, erreur, taylor = approximer(f_str, x_val, ordre, a)
        
        if exact is not None:
            print(f"{ordre:8d} {approx:15.10f} {exact:15.10f} {erreur:12.2e}")
            resultats.append({
                'Fonction': f_str,
                'x': x_val,
                'Ordre': ordre,
                'Approx': approx,
                'Exact': exact,
                'Erreur': erreur
            })
    
    print()

# e) Tableau comparatif
print("e) Tableau comparatif\n")

if resultats:
    df = pd.DataFrame(resultats)
    # Formater l'affichage
    pd.set_option('display.float_format', lambda x: f'{x:.6f}' if abs(x) < 1 else f'{x:.2e}')
    print(df.to_string(index=False))
    print("\nâœ… Tableau crÃ©Ã© avec succÃ¨s !")
    print("\nğŸ’¡ Observations :")
    print("   - L'erreur dÃ©croÃ®t exponentiellement avec l'ordre")
    print("   - sin et cos convergent trÃ¨s vite")
    print("   - exp converge rapidement pour x petit")
    print("   - log(1+x) a besoin d'ordres plus Ã©levÃ©s prÃ¨s de 0")
```
